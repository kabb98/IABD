## Enviar a Kafka

```python

from random import choice
from confluent_kafka import Producer
import time
```

```python
def callback(err, msg):
    if err is not None:
        print("Failed to deliver message: %s: %s" % (str(msg), str(err)))
    else:
        print("Message produced: %s" % (str(msg)))
```

```python
import json
```

```python
def main():
    if __name__ == '__main__':

        config = {
            # User-specific properties that you must set
            'bootstrap.servers': 'localhost:9092',

            # Fixed properties
            'acks': 'all'
        }

        producer: Producer = Producer(config)

        topic = 'bitcoin'

        while True:
            # Load the last price of the api of bitcoin
            response = requests.get(url_bitcoin).json()

            # Get the price of bitcoin
            price_usd = response['bpi']['USD']['rate']
            price_gbp = response['bpi']['GBP']['rate']
            price_eur = response['bpi']['EUR']['rate']

            # Get the date of the price
            date = response['time']['updated']

            # Create the dict -> date: price_currencies
            message = {
                'date': date,
                'price_usd': price_usd
            }

            # Send the message to the topic
            producer.produce(topic, key=str(price_usd), value=json.dumps(message), callback=callback)

            message = {
                'date': date,
                'price_eur': price_eur
            }
            producer.produce(topic, key=str(price_eur), value=json.dumps(message), callback=callback)

            message = {
                'date': date,
                'price_gbp': price_gbp
            }
            producer.produce(topic, key=str(price_gbp), value=json.dumps(message), callback=callback)

            # Wait 5 seconds
            time.sleep(5)

            producer.flush()

```

```python

if __name__ == '__main__':
    main()
```
