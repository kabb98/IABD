```python
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

### 1. ¿Cómo se distribuyen las longitudes del pico (bill_length_mm) y de la aleta (flipper_length_mm) entre las diferentes especies?

```python
sns.violinplot(x='species', y='bill_length_mm', data=df, hue='species')
```

```python
sns.violinplot(x='species', y='flipper_length_mm', data=df, hue='species')
```

### 2. ¿Qué diferencias existen en la masa corporal (body_mass_g) entre las tres especies de pingüinos?

```python
sns.boxplot(x='species', y='body_mass_g', data=df, hue='species')
```

### 3. ¿Existe una correlación entre bill_length_mm y bill_depth_mm? ¿Cómo varía entre las especies? 

```python
g = sns.scatterplot(x='bill_length_mm', y='flipper_length_mm',
                    data=df, hue='species', style='species')
```

### 4. ¿Cuál es la relación entre la longitud de la aleta (flipper_length_mm) y la masa corporal (body_mass_g)? ¿Se comporta igual en todas las especies?

```python
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=df, hue='species', style='species')
```

### 5. ¿Cómo varía la distribución de especies entre las islas (island)? ¿Alguna especie es exclusiva de una isla?

```python
sns.countplot(x='island', hue='species', data=df)
```

### 6. ¿Existen diferencias significativas en el tamaño del pico (bill_length_mm, bill_depth_mm) entre machos y hembras?

```python
# Gráfico de caja para bill_length_mm
sns.boxplot(x='sex', y='bill_length_mm', data=df, hue='sex')
```

```python

# Gráfico de caja para bill_depth_mm
sns.boxplot(x='sex', y='bill_depth_mm', data=df, hue='sex')
```

### 7. ¿Cómo afecta la isla de origen (island) a la masa corporal (body_mass_g) y a la longitud de la aleta (flipper_length_mm)? 

```python
# Gráfico de caja para body_mass_g
sns.boxplot(x='island', y='body_mass_g', data=df, hue='island')

# Gráfico de caja para flipper_length_mm
sns.boxplot(x='island', y='flipper_length_mm', data=df, hue='island')
```

### 8. ¿Qué especie es la más común en el dataset? ¿Hay variaciones significativas en las frecuencias entre las islas?

```python
import matplotlib.pyplot as plt
```

```python
# Graficar la frecuencia de cada especie
sns.countplot(x='species', data=df, palette='viridis')

# Mostrar el título del gráfico
plt.title('Frecuencia de cada especie en el dataset')

# Mostrar el gráfico
plt.show()
```

```python
sns.countplot(x='island', hue='species', data=df, palette='viridis')

# Mostrar el título del gráfico
plt.title('Distribución de especies entre las islas')

# Mostrar el gráfico
plt.show()
```

### 9. ¿Cómo se diferencian las especies cuando combinamos variables como bill_length_mm y flipper_length_mm?

```python

sns.scatterplot(x='bill_length_mm', y='flipper_length_mm', data=df, hue='species', style='species')
plt.title('Relación entre bill_length_mm y flipper_length_mm por especie')
plt.show()
```
