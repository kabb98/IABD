```python
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

### 1. ¿Cómo se distribuyen las longitudes del pico (bill_length_mm) y de la aleta (flipper_length_mm) entre las diferentes especies?

```python
sns.violinplot(x='species', y='bill_length_mm', data=df, hue='species')
```

```python
sns.violinplot(x='species', y='flipper_length_mm', data=df, hue='species')
```

### 2. ¿Qué diferencias existen en la masa corporal (body_mass_g) entre las tres especies de pingüinos?

```python
sns.boxplot(x='species', y='body_mass_g', data=df, hue='species')
```

```python
g = sns.scatterplot(x='bill_length_mm', y='flipper_length_mm',
                    data=df, hue='species', style='species')
```
