```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set_theme()
sns.relplot(x='total_bill', y='tip', data=tips, col='time', hue='smoker', style='smoker', size='size')
```

```python
sns.set_theme()
sns.relplot(x='total_bill', y='tip', data=tips, col='time', hue='sex', style='sex', size='size')
```

```python
sns.set_theme()
sns.relplot(x='total_bill', y='tip', data=tips, col='time', hue='day', style='day', size='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, col='day',
            hue='smoker', style='smoker', size='size')
```

```python
dots = sns.load_dataset('dots')
```

```python
sns.relplot(kind="line" ,x='time', y='firing_rate', data=dots, col='align', size='coherence', style='choice', hue='choice')
```

```python
sns.relplot(kind="line" ,x='time', y='firing_rate', data=dots, col='align',  style='choice', hue='choice', facet_kws={'sharex': True})
```
