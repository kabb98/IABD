## Importamos las librerías necesarias

```python
import pandas as pd
```

## Leemos el fichero CVS

```python
df = pd.read_csv('./country_vaccinations.csv')
```

```python
df.head()
```

## Promedio de dosis administradas por país

```python
pd.options.display.float_format = '{:,.2f}'.format
avg_doses = df.groupby('country')['total_vaccinations'].mean()
```

```python
avg_doses.head()
```

## Tasa de vacunación por cada 100000 habitantes de cada país

```python
tasa_vacunacion = df.groupby('country')['total_vaccinations_per_hundred'].sum() * 100
```
